# ODEs: Explicit vs Implicit Methods

In this notebook I will show several examples of explicit and implicit solvers, and the way they are different and similar to each other, and touch on concepts of accuracy, convergence & stability.

N.B. For a Jupyter notebook, In order to execute the code in a cell, click on the cell, then type Shift+Enter on your keyboard

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as scp
%matplotlib inline
```

Let us start with a very simple differential equation. Remember that a differential equation tries to solve for the relationship between two variables, in this case $x$ and $y$:

$y^\prime = x$, with $y(0)=0$.

The analytical solution to this equation is $y(x)=\frac{1}{2} x^2$.



## Explicit Euler and Improved Euler

Let's assume we don't know the analysitcal solution to the previous equation. To find out the value of $y$ at a certain value of $x$, we need to somehow get to the value of $x$ we want by making jumps. This is numerically solving the ODE.

The most simple method we can think of is to approximate $y^\prime$ as $[y(x+\Delta x) - x(t)]/\Delta x$, and thus to write a numerical scheme as follows:

$y(x+\Delta x) = y(t) + y^\prime \Delta x$.

But how do we evaluate $y^\prime$? The simplest thing to do is to evaluate $y^\prime$ at a place we know it, like $x$.

Thus, the simplest method we can write, the explicit Euler looks like:

$y(x+\Delta x) = y(x) + x \Delta x$.

Let us implement this scheme for the above method, to try to find out $y$ up to $x=2$. We will first pick $\Delta x=0.1$

```python
DeltaX=0.2
xend = 2.0
x0=0.
y0=0.

Number_Steps = int(xend/DeltaX)
xs=[]
ys=[]

ys.append(y0)
xs.append(x0)
for i in range(Number_Steps):
    x_old=xs[-1]
    y_old=ys[-1]
    x_new=x_old+DeltaX
    y_new=y_old+DeltaX*x_old
    xs.append(x_new)
    ys.append(y_new)
```

Let us compare the solution we just obtained to the analytical solution.

```python
plt.plot(xs,ys,'.-')
y_analytical=0.5*(np.array(xs)**2.)
plt.plot(xs,y_analytical,'.-')
```

Whata about the error we are incurring in?

```python
Relative_Error=np.abs(y_analytical-ys)/np.abs(ys)
Absolute_Error=np.abs(y_analytical-ys)

plt.plot(xs,Absolute_Error,'.-')

Relative_Error[-1]*100.
```

As you can see, the error grows with every step we are taking. The error at our last point is already $11\%$.

What would happen if we take smaller steps? Let us try $\Delta x=0.1$

```python
DeltaX=0.1
xend = 2.0
x0=0.
y0=0.

Number_Steps = int(xend/DeltaX)
xs=[]
ys=[]

ys.append(y0)
xs.append(x0)
for i in range(Number_Steps):
    x_old=xs[-1]
    y_old=ys[-1]
    x_new=x_old+DeltaX
    y_new=y_old+DeltaX*x_old
    xs.append(x_new)
    ys.append(y_new)
```

```python
plt.plot(xs,ys,'.-')
y_analytical=0.5*(np.array(xs)**2.)
plt.plot(xs,y_analytical,'.-')
np.abs(y_analytical[-1]-ys[-1])/np.abs(ys[-1])
```

By taking steps half the size, we have reduced our error by a half. This is because the Euler method is a first order method, and the error is proportional to $\Delta t$. 

We can estimate that to bring the error below $1%$, we would need to divide the time-step further, by a factor 5, that is, down to $\Delta t=0.02$.

```python
DeltaX=0.02
xend = 2.0
x0=0.
y0=0.

Number_Steps = int(xend/DeltaX)
xs=[]
ys=[]

ys.append(y0)
xs.append(x0)
for i in range(Number_Steps):
    x_old=xs[-1]
    y_old=ys[-1]
    x_new=x_old+DeltaX
    y_new=y_old+DeltaX*x_old
    xs.append(x_new)
    ys.append(y_new)
```

```python
plt.plot(xs,ys,'.-')
y_analytical=0.5*(np.array(xs)**2.)
plt.plot(xs,y_analytical,'.-')

np.abs(y_analytical[-1]-ys[-1])/np.abs(ys[-1])
```

We have finally achieved $1\%$ error. However, by doing so, we have increased substantially the number of timesteps up to 100, from the initial 10. This has made the runtime 10 times slower, and has increased our memory requirements by a factor 10. If we wanted to achieve $0.1\%$ precision, we would need to go down to $\Delta x=0.002$, and multiply again our requirement. 

We could use a higher order method, whose error is $\Delta x^n$, where $n$ is the order of the method. The most popular methods are known as the Runge-Kutta method, and these are usually 4th or higher order, and they are used in the ode45 routines in matlab (or Python). These also have variable time-steps and perform error checking. I will leave these for later, and stay with fixed and constant steps for now.

For you to play around, in the following code, I have implemented a second order method known as the improved Euler, or Heun's method. This method is described in wikipedia:

https://en.wikipedia.org/wiki/Heun's_method

```python
DeltaX=0.1
xend = 2.0
x0=0.
y0=0.

Number_Steps = int(xend/DeltaX)
xs=[]
ys=[]

ys.append(y0)
xs.append(x0)
for i in range(Number_Steps):
    x_old=xs[-1]
    y_old=ys[-1]
    x_new=x_old+DeltaX
    m1=x_old
    m2=x_new    
    y_new=y_old+DeltaX*0.5*(m1+m2)
    xs.append(x_new)
    ys.append(y_new)
```

```python
plt.plot(xs,ys,'.-')
y_analytical=0.5*(np.array(xs)**2.)
plt.plot(xs,y_analytical,'.-')

np.abs(y_analytical[-1]-ys[-1])/np.abs(ys[-1])
```

As you can see, the error is now zero to machine precision. Now this is cheating a bit because our solution is a second order polynomial, so we can get it exactly with a second order scheme.

Let's try a slightly harder equation:

$y^\prime = cos(x)$, with $y(0)=0$.

The solution for this equation is $y=sin(x)$.

Let us first try Euler with $\Delta x =0.1$

```python
DeltaX=0.1
xend = 2.0
x0=0.
y0=0.

Number_Steps = int(xend/DeltaX)
xs=[]
ys=[]

ys.append(y0)
xs.append(x0)
for i in range(Number_Steps):
    x_old=xs[-1]
    y_old=ys[-1]
    x_new=x_old+DeltaX
    y_new=y_old+DeltaX*np.cos(x_old)
    xs.append(x_new)
    ys.append(y_new)
    
plt.plot(xs,ys,'.-')
y_analytical=np.sin(xs)
plt.plot(xs,y_analytical,'.-')

np.abs(y_analytical[-1]-ys[-1])/np.abs(ys[-1])
```

And now Heun's method. You will see that the error is much smaller.

```python
DeltaX=0.1
xend = 2.0
x0=0.
y0=0.

Number_Steps = int(xend/DeltaX)
xs=[]
ys=[]

ys.append(y0)
xs.append(x0)
for i in range(Number_Steps):
    x_old=xs[-1]
    y_old=ys[-1]
    x_new=x_old+DeltaX
    m1=np.cos(x_old)
    m2=np.cos(x_new)
    y_new=y_old+DeltaX*0.5*(m1+m2)
    xs.append(x_new)
    ys.append(y_new)

plt.plot(xs,ys,'.-')
y_analytical=np.sin(xs)
plt.plot(xs,y_analytical,'.-')

np.abs(y_analytical[-1]-ys[-1])/np.abs(ys[-1])    
```

If we double the time-step, the error goes down by a factor four, as it is a second-order method.

```python
DeltaX=0.05
xend = 2.0
x0=0.
y0=0.

Number_Steps = int(xend/DeltaX)
xs=[]
ys=[]

ys.append(y0)
xs.append(x0)
for i in range(Number_Steps):
    x_old=xs[-1]
    y_old=ys[-1]
    x_new=x_old+DeltaX
    m1=np.cos(x_old)
    m2=np.cos(x_new)
    y_new=y_old+DeltaX*0.5*(m1+m2)
    xs.append(x_new)
    ys.append(y_new)

plt.plot(xs,ys,'.-')
y_analytical=np.sin(xs)
plt.plot(xs,y_analytical,'.-')

np.abs(y_analytical[-1]-ys[-1])/np.abs(ys[-1])    
```

So to reduce our error, we can use either a higher order method, which performs more calculations per step, but allows us to take larger steps, or use a smaller step. 

There are also other things that can happen if time-step is too large. Our method could also destabilize, or not converge to the correct result. I will not talk about these things here, but there are other things to think about.

As Wikipedia says:

Three central concepts in this analysis are:

1. Convergence: whether the method approximates the solution,
2. Order: how well it approximates the solution, and
3. Stability: whether errors are damped out.

We will revisit (1) and (3) later. They sometimes go together, they sometimes don't.

## Implicit (backwards) Euler and Stiff Equations

Up to now, we have only treated explicit methods, that is, methods that use information which is available to us already.

We could think of writing a scheme for solving:

$y^\prime = f(x,y)$,

which not only uses information from the current step we are at (or from steps backwards), but also information from the future. However, this is not known, and if the equation is not linear the method to obtain these is quite complicated. You can check the book to learn more. 

For now let us try to solve the following equation:

$y^\prime = -100 y$ with $y(0)=1$,

whose analytical solution is $y=\exp(-100x)$.


What happens if we try a first order Euler?

```python
DeltaX=0.1
xend = 0.5
x0=0.
y0=1.

Number_Steps = int(xend/DeltaX)
xs=[]
ys=[]

ys.append(y0)
xs.append(x0)
for i in range(Number_Steps):
    x_old=xs[-1]
    y_old=ys[-1]
    x_new=x_old+DeltaX
    y_new=y_old+DeltaX*-100*y_old
    xs.append(x_new)
    ys.append(y_new)
    
plt.plot(xs,ys,'.-')
#y_analytical=np.sin(xs)
#plt.plot(xs,y_analytical,'.-')

np.abs(y_analytical[-1]-ys[-1])/np.abs(ys[-1])
```

The solution is 6e5 by the end of the run. The solution has not only failed to approach the correct value, but it is spiraling out of control! This means that it has become unstable. 

Let us reduce the step size.

```python
DeltaX=0.005
xend = 0.5
x0=0.
y0=1.

Number_Steps = int(xend/DeltaX)
xs=[]
ys=[]

ys.append(y0)
xs.append(x0)
for i in range(Number_Steps):
    x_old=xs[-1]
    y_old=ys[-1]
    x_new=x_old+DeltaX
    y_new=y_old+DeltaX*-100.*y_old
    xs.append(x_new)
    ys.append(y_new)
    
plt.plot(xs,ys,'.-')
y_analytical=np.exp(-100.*np.array(xs))
plt.plot(xs,y_analytical,'.-')

#np.abs(y_analytical[-1]-ys[-1])/np.abs(ys[-1])
```

By reducing the time-step enough, we manage to stabilize the method, but the error is still very large. If we show the grpah in semi-log scale, we see that the error grows quite fast.

```python
plt.semilogy(xs,ys,'.-')
y_analytical=np.exp(-100.*np.array(xs))
plt.semilogy(xs,y_analytical,'.-')
```

If we change the equation now to $y^\prime=-1000y$ with the same initial conditions, our time-step has to go down even further to stabilize the method, and to reach reasonable results. Try to find out a good time-step to have less than $1\%$ error by the end of the run.

```python
DeltaX=0.0005
xend = 0.1
x0=0.
y0=1.

Number_Steps = int(xend/DeltaX)
xs=[]
ys=[]

ys.append(y0)
xs.append(x0)
for i in range(Number_Steps):
    x_old=xs[-1]
    y_old=ys[-1]
    x_new=x_old+DeltaX
    y_new=y_old+DeltaX*-1000.*y_old
    xs.append(x_new)
    ys.append(y_new)
    
plt.semilogy(xs,ys,'.-')
y_analytical=np.exp(-1000.*np.array(xs))
plt.semilogy(xs,y_analytical,'.-')

#np.abs(y_analytical[-1]-ys[-1])/np.abs(ys[-1])
```

We can also try to solve this using an implicit method. Let us try the backwards Euler:

$y(t+\Delta t) = y(t) + \Delta t f(y(t+\Delta t), t+\Delta t)$

Here we can implement an implicit method quite easily, because $f(y,t)=-1000y$, and the equation is linear. So we can just shift everything to the left hand side, and our method looks like:

$y(t+\Delta t)(1+1000*\Delta t) = y(t)$


```python
DeltaX=0.1
xend = 0.5
x0=0.
y0=1.

Number_Steps = int(xend/DeltaX)
xs=[]
ys=[]

ys.append(y0)
xs.append(x0)
for i in range(Number_Steps):
    x_old=xs[-1]
    y_old=ys[-1]
    x_new=x_old+DeltaX
    y_new=y_old/(1.+1000.*DeltaX)
    xs.append(x_new)
    ys.append(y_new)
    
plt.plot(xs,ys,'.-')
y_analytical=np.exp(-1000.*np.array(xs))
plt.plot(xs,y_analytical,'.-')
```

Our method is now stable for very large timesteps! You can go back to the Euler method and check that it doesn't stabilize until you have very small values of $\Delta t$. The error in our case might still be large, but we can improve on it by increasing $\Delta t$ a bit, but not to the levels we were doing before. Let us check $\Delta x=0.005$

```python
DeltaX=0.0005
xend = 0.1
x0=0.
y0=1.

Number_Steps = int(xend/DeltaX)
xs=[]
ys=[]

ys.append(y0)
xs.append(x0)
for i in range(Number_Steps):
    x_old=xs[-1]
    y_old=ys[-1]
    x_new=x_old+DeltaX
    y_new=y_old/(1.+100.*DeltaX)
    xs.append(x_new)
    ys.append(y_new)
    
plt.plot(xs,ys,'.-')
y_analytical=np.exp(-100.*np.array(xs))
plt.plot(xs,y_analytical,'.-')
```

The match is very decent. This is what is called a stiff ODE, and shows why implicit methods are sometimes used even if it can be hard to write them down. They are usually more stable and can take larger time-steps. 

Usually implicit methods are 2nd order, much lower than RK methods. Ode13s in matlab is a good example of a method used for implicit equations. Backwards differentiation methods or implicit Runge Kutta are the most common methods in these categories.
